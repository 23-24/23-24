```python
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np
import os

# Load the pre-trained ResNet50 model
model = ResNet50(weights='imagenet', include_top=False, pooling='avg')
```

```python
output_file = 'features.npy'

# Initialize an empty list to store the features
features_list = []

# Loop over each file in the input directory
for filename in os.listdir(output_dir):
    # Load the image file and convert it to a numpy array
    img_path = os.path.join(output_dir, filename)
    img = load_img(img_path, target_size=(224, 224))
    img_array = img_to_array(img)
    
    # Preprocess the image array
    img_array = preprocess_input(img_array)
    
    # Add a batch dimension to the array
    img_array = np.expand_dims(img_array, axis=0)
    
    # Use the ResNet50 model to extract features from the image
    features = model.predict(img_array)
    features_list.append(features)

# Convert the list of features to a numpy array and save it to a file
features_array = np.concatenate(features_list, axis=0)
np.save(output_file, features_array)
```

```python


# Read captions from text document
with open('captions.txt', 'r') as f:
    captions = f.readlines()

# Preprocess captions
tokenizer = Tokenizer()
tokenizer.fit_on_texts(captions)
word_index = tokenizer.word_index
sequences = tokenizer.texts_to_sequences(captions)
padded_sequences = pad_sequences(sequences, padding='post')

# Save preprocessed captions to numpy array file
np.save('captions.npy', padded_sequences)
```

```python
# Load features and captions
features = np.load('features.npy')
captions = np.load('captions.npy')

# Determine the number of samples
num_samples = len(features)

# Shuffle the indices
indices = np.arange(num_samples)
np.random.shuffle(indices)

# Split the data into train, validation, and test sets
train_indices = indices[:int(0.7*num_samples)]
val_indices = indices[int(0.7*num_samples):int(0.85*num_samples)]
test_indices = indices[int(0.85*num_samples):]

train_features = features[train_indices]
train_captions = captions[train_indices]
val_features = features[val_indices]
val_captions = captions[val_indices]
test_features = features[test_indices]
test_captions = captions[test_indices]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, GRU, Dense

# Define the maximum length of the captions and the size of the vocabulary
max_caption_len = 20
vocab_size = 5000

# Define the embedding size and the hidden size of the LSTM/GRU layer
embedding_size = 256
hidden_size = 512
# Define the model architecture
model = Sequential()
model.add(Embedding(vocab_size, embedding_size, input_length=max_caption_len))
model.add(LSTM(hidden_size))
model.add(Dense(hidden_size, activation='relu'))
model.add(Dense(vocab_size, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam')

# Print the model summary
model.summary()
```
